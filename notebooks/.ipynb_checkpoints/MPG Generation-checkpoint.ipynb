# MPG Generation
## 1. Distributions 
- Let $\mathcal{G}$ be a distribution of graphs
- Let $\mathcal{D}(u,v)$ be a distribution of weights depending on $u,v\in V$

An MPG $M$ following the distribution $\mathcal{G}\vartriangle \mathcal{D}$ if:
- The underlying graph of $M$ follows the distribution $\mathcal{G}$
- The weight of each edge $(u,v)$ follow the distribution $\mathcal{D}(u,v)$:
    $$
    \forall u,v\in E,\quad L(u,v)\sim \mathcal{D}(u,v)
    $$
