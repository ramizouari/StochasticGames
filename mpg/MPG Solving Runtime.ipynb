```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import time,date,datetime
import games.mpg as mpg
import networkx as nx
import timeit
```

```python
import os
import re
base_dir="/run/media/ramizouari/INTENSO/MPG/"
directory="/run/media/ramizouari/INTENSO/MPG/dataset"
dataset=[]
k=0
for file in os.listdir(directory):
    match=re.search(R"gnp_uniform_mpg_(\d+)_(\d+)_(\d+)(\.edgelist\.gz)",file)
    if match:
        n=int(match.group(1))
        c=int(match.group(2))
        extension=match.group(4)
    if c==3 and n==100:
        print(n)
        graph=nx.read_weighted_edgelist(os.path.join(directory,file),create_using=mpg.MeanPayoffGraph,nodetype=int)
        for e in graph.edges:
            for u in graph.edges[e]:
                graph.edges[e][u]=int(graph.edges[e][u])
        start=timeit.default_timer()
        execution_start_time=datetime.now()
        graph=mpg.optimal_strategy_pair(graph,method="ACOS")
        end=timeit.default_timer()
        execution_end_time=datetime.now()
        dataset.append(dict(time=end-start,filename=file,execution_start=execution_start_time,execution_end=execution_end_time))
dataset=pd.DataFrame(dataset,columns=["filename","execution_start","execution_end","time"])
```
